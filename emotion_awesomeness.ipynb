```python
anger_list = []
contempt_list = []
fear_list = []
disgust_list = []
neutral_list = []
sadness_list = []
surprise_list = []
```

```python
for i in scores:
    anger_list.append(i['anger'])
    contempt_list.append(i['contempt'])
    fear_list.append(i['fear'])
    disgust_list.append(i['disgust'])
    neutral_list.append(i['neutral'])
    sadness_list.append(i['sadness'])
    surprise_list.append(i['surprise'])
```

```python
anger_avg = sum(anger_list)/len(anger_list)
contempt_avg = sum(contempt_list)/len(contempt_list)
fear_avg = sum(fear_list)/len(fear_list)
disgust_avg = sum(disgust_list)/len(disgust_list)
neutral_avg = sum(neutral_list)/len(neutral_list)
sadness_avg = sum(sadness_list)/len(sadness_list)
surprise_avg = sum(surprise_list)/len(surprise_list)
```
